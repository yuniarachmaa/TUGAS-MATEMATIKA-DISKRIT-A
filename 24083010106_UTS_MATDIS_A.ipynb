# **ABSTRAK**
Stunting  adalah  masalah  kesehatan  serius  yang  berdampak  jangka  panjang  pada  pertumbuhan anak-anak,  terutama  di  wilayah  Cimacan.  Pengabdian  kepada  Masyarakat  (PkM)  ini  bertujuan  untuk mengevaluasi efektivitas pendidikan kesehatan dalam meningkatkan pengetahuan wanita usia subur (WUS) tentang  stunting.  Penelitian  ini  menggunakan  metode  penyuluhan  kesehatan  yang  melibatkan  16  peserta WUS, dengan waktu pelaksaan sekitar satu jam. Penyuluhan berlangsung dalam tiga sesi: pengisian pre-test, pemaparan   materi   dan   diskusi,   serta   pengisian   kuosioner   post-test.   Data   dianalisis   dengan   uji   t-test berpasangan untuk mengukur perubahan skor pengetahuan. Hasil penelitian menunjukkan peningkatan skor rata-rata   dari   pre-test   (M=71,875;   SD=16,419)   ke   post-test   (M=83,125;   SD=9,465).   Analisis   statistik menunjukkan  hasil  yang  signifikan  (p=0,017  <  0,05),  yang  menunjukkan  peningkatan  pemahaman  peserta mengenai   stunting.   Kesimpulannya,   pendidikan   kesehatan   berkontribusi   positif   terhadap   peningkatan kesadaran  WUS  tentang  stunting.  Temuan  ini  menekankan  pentingnya  edukasi  yang  efektif  dalam  upaya pencegahan stunting.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from IPython.display import display, HTML  # menampilkan tabel visual

# ============================================================
# Analisis Jumlah Vokal dan Konsonan (Bagian Abstrak)
# ============================================================

# ==============================
# 1. Baca dan siapkan teks
# ==============================
with open("abstrak_bersih.txt", "r", encoding="utf-8") as f:
    teks = f.read()

# Ubah huruf jadi lowercase dan bersihkan
teks = teks.lower()
teks = re.sub(r'[^a-z\s]', '', teks)
teks = re.sub(r'\s+', ' ', teks).strip()

# Pisahkan menjadi daftar kata
kata_list = teks.split()

# ==============================
# 2. Hitung jumlah vokal dan konsonan per kata
# ==============================
data = []
for kata in kata_list:
    vokal = sum(1 for c in kata if c in "aiueo")
    konsonan = sum(1 for c in kata if c.isalpha() and c not in "aiueo")
    data.append({"kata": kata, "vokal": vokal, "konsonan": konsonan})

df = pd.DataFrame(data)

# ==============================
# 3. Pivot 1 - Frekuensi kombinasi konsonanâ€“vokal
# ==============================
pivot1 = pd.pivot_table(
    df,
    index="vokal",           # sumbu Y
    columns="konsonan",      # sumbu X
    values="kata",
    aggfunc="count",
    fill_value=0
)

print("=== PIVOT 1: FREKUENSI KOMBINASI KONSONANâ€“VOKAL (Abstrak) ===\n")
display(HTML(pivot1.to_html(classes='table table-striped table-bordered', border=0)))

# Heatmap 1
plt.figure(figsize=(8,6))
sns.heatmap(pivot1, annot=True, fmt="d", cmap="Blues")
plt.title("Frekuensi Kombinasi Vokalâ€“Konsonan (Abstrak)")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()

# ==============================
# 4. Pivot 2 - Cacah kata per kombinasi
# ==============================
df_grouped = df.groupby(["vokal", "konsonan"])["kata"].nunique().reset_index(name="jumlah_kata")

pivot2 = pd.pivot_table(
    df_grouped,
    index="vokal",
    columns="konsonan",
    values="jumlah_kata",
    fill_value=0
)

print("\n=== PIVOT 2: CACAH KATA PER KOMBINASI (Abstrak) ===")
display(HTML(pivot2.to_html(classes='table table-striped table-bordered', border=0)))

# Heatmap 2
plt.figure(figsize=(8,6))
sns.heatmap(pivot2, annot=True, fmt=".0f", cmap="Greens")
plt.title("Cacah Kata per Kombinasi Vokalâ€“Konsonan (Abstrak)")
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()

# ==============================
# 5. Statistik Tambahan
# ==============================
jumlah_kata = len(df)
rata_vokal = round(df['vokal'].mean(), 2)
rata_konsonan = round(df['konsonan'].mean(), 2)

print("\n=== STATISTIK TAMBAHAN (Abstrak) ===")
print("Jumlah total kata dalam abstrak:", jumlah_kata)
print("Rata-rata jumlah vokal per kata:", rata_vokal)
print("Rata-rata jumlah konsonan per kata:", rata_konsonan)
```

* Kombinasi Konsonan-Vokal dalam kata

Hasil tabel pivot menunjukkan bahwa kombinasi huruf dengan jumlah vokal antara 2â€“4 dan konsonan antara 3â€“6 memiliki frekuensi tertinggi. Pada heatmap berwarna biru, daerah tersebut tampak paling gelap, menandakan konsentrasi kata yang tinggi di kombinasi tersebut. Artinya, sebagian besar kata dalam abstrak terdiri atas huruf vokal dan konsonan yang seimbang. Pola ini menegaskan bahwa teks abstrak didominasi oleh kata-kata berukuran sedang, yang umum digunakan dalam tulisan ilmiah untuk menjaga kejelasan dan keterbacaan.

```python
# ============================================================
# Analisis Jumlah Kata per Kalimat (Bagian Abstrak)
# ============================================================

# ==============================
# 1. Baca & Bersihkan Teks
# ==============================
with open("abstrak_bersih.txt", "r", encoding="utf-8") as f:
    teks = f.read()

# Bersihkan teks dari spasi ganda & karakter aneh
teks = re.sub(r'\s+', ' ', teks.strip())

# ==============================
# 2. Pisahkan Kalimat
# ==============================
# Pisahkan berdasarkan tanda titik, tanda seru, atau tanda tanya
kalimat_list = re.split(r'[.!?]+', teks)
kalimat_list = [kal.strip() for kal in kalimat_list if kal.strip()]

# ==============================
# 3. Hitung Jumlah Kata per Kalimat
# ==============================
data = []
for i, kalimat in enumerate(kalimat_list, start=1):
    kata = kalimat.split()
    jumlah_kata = len(kata)
    data.append({
        "Kalimat_ke": i,
        "Teks_Kalimat": kalimat,
        "Jumlah_Kata": jumlah_kata
    })

df_kalimat = pd.DataFrame(data)

# ==============================
# 4. Tampilkan Tabel Visual
# ==============================
print("=== TABEL JUMLAH KATA PER KALIMAT (Abstrak) ===\n")
display(HTML(df_kalimat.to_html(classes='table table-striped table-bordered', index=False)))

# ==============================
# 5. Visualisasi Bar Chart
# ==============================
plt.figure(figsize=(8,5))
sns.barplot(
    data=df_kalimat,
    x="Kalimat_ke",
    y="Jumlah_Kata",
    hue="Kalimat_ke",
    dodge=False,
    legend=False,
    palette="RdPu"
)
plt.title("Jumlah Kata per Kalimat (Abstrak)", fontsize=12, color="#d63384")
plt.xlabel("Kalimat ke-", fontsize=10)
plt.ylabel("Jumlah Kata", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ==============================
# 6. Statistik Tambahan
# ==============================
jumlah_kalimat = len(df_kalimat)
rata_rata = round(df_kalimat["Jumlah_Kata"].mean(), 2)
maks = df_kalimat["Jumlah_Kata"].max()
minim = df_kalimat["Jumlah_Kata"].min()

print("\n=== STATISTIK TAMBAHAN (Abstrak) ===")
print(f"Jumlah total kalimat: {jumlah_kalimat}")
print(f"Rata-rata jumlah kata per kalimat: {rata_rata}")
print(f"Kalimat terpanjang: {maks} kata")
print(f"Kalimat terpendek: {minim} kata")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from IPython.display import display, HTML  # menampilkan tabel visual

# ============================================================
# Analisis Jumlah Vokal dan Konsonan (Bagian Pendahuluan)
# ============================================================

# ==============================
# 1. Baca & Bersihkan Teks
# ==============================
with open("pendahuluan_bersih.txt", "r", encoding="utf-8") as f:
    teks = f.read()

# Lowercase & bersihkan karakter
teks = teks.lower()
teks = re.sub(r'[^a-z\s]', '', teks)
teks = re.sub(r'\s+', ' ', teks).strip()

# Pisahkan kata
kata_list = teks.split()

# ==============================
# 2. Hitung jumlah vokal & konsonan per kata
# ==============================
data = []
for kata in kata_list:
    vokal = sum(1 for c in kata if c in "aiueo")
    konsonan = sum(1 for c in kata if c.isalpha() and c not in "aiueo")
    data.append({"kata": kata, "vokal": vokal, "konsonan": konsonan})

df = pd.DataFrame(data)

# ==============================
# 3. Pivot 1 - Frekuensi kombinasi konsonanâ€“vokal
# ==============================
pivot1 = pd.pivot_table(
    df,
    index="vokal",           # sumbu Y
    columns="konsonan",      # sumbu X
    values="kata",
    aggfunc="count",
    fill_value=0
)

print("=== PIVOT 1: FREKUENSI KOMBINASI KONSONANâ€“VOKAL (Pendahuluan) ===\n")
display(HTML(pivot1.to_html(classes='table table-striped table-bordered', border=0)))

# Heatmap 1
plt.figure(figsize=(10,7))
sns.heatmap(pivot1, annot=True, fmt="d", cmap="Blues")
plt.title("Frekuensi Kombinasi Vokalâ€“Konsonan (Pendahuluan)", fontsize=12)
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()

# ==============================
# 4. Pivot 2 - Cacah kata per kombinasi
# ==============================
df_grouped = df.groupby(["vokal", "konsonan"])["kata"].nunique().reset_index(name="jumlah_kata")

pivot2 = pd.pivot_table(
    df_grouped,
    index="vokal",
    columns="konsonan",
    values="jumlah_kata",
    fill_value=0
)

print("\n=== PIVOT 2: CACAH KATA PER KOMBINASI (Pendahuluan) ===")
display(HTML(pivot2.to_html(classes='table table-striped table-bordered', border=0)))

# Heatmap 2
plt.figure(figsize=(10,7))
sns.heatmap(pivot2, annot=True, fmt=".0f", cmap="Greens")
plt.title("Cacah Kata per Kombinasi Vokalâ€“Konsonan (Pendahuluan)", fontsize=12)
plt.xlabel("Jumlah Konsonan")
plt.ylabel("Jumlah Vokal")
plt.show()

# ==============================
# 5. Statistik Tambahan
# ==============================
jumlah_kata = len(df)
rata_vokal = round(df['vokal'].mean(), 2)
rata_konsonan = round(df['konsonan'].mean(), 2)

print("\n=== STATISTIK TAMBAHAN (Pendahuluan) ===")
print("Jumlah total kata dalam pendahuluan:", jumlah_kata)
print("Rata-rata jumlah vokal per kata:", rata_vokal)
print("Rata-rata jumlah konsonan per kata:", rata_konsonan)
```
