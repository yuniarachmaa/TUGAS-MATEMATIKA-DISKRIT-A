```python
import pandas as pd

def soal1(n):
    lhs = sum(2*i - 1 for i in range(1, n+1))
    rhs = n**2
    return lhs, rhs, lhs == rhs

def soal3(n):
    lhs = sum(i**2 for i in range(1, n+1))
    rhs = n*(n+1)*(2*n+1)//6
    return lhs, rhs, lhs == rhs

# Buat tabel untuk n=1 sampai 15
data1 = []
data3 = []

for n in range(1, 16):
    lhs1, rhs1, valid1 = soal1(n)
    data1.append([n, lhs1, rhs1, valid1])

    lhs3, rhs3, valid3 = soal3(n)
    data3.append([n, lhs3, rhs3, valid3])

# DataFrame
df1 = pd.DataFrame(data1, columns=["n", "LHS (jumlah ganjil)", "RHS (n^2)", "Valid?"])
df3 = pd.DataFrame(data3, columns=["n", "LHS (jumlah kuadrat)", "RHS (rumus)", "Valid?"])

print("=== Soal (1): Jumlah bilangan ganjil ===")
display(df1)

print("\n=== Soal (3): Jumlah kuadrat bilangan ===")
display(df3)
```

# Memeriksa soal persamaan (1) :
1+3+5+⋯+(2n−1)=n^2

Hasil tabel menunjukkan bahwa untuk setiap nilai 𝑛 dari 1 sampai 15, sisi kiri (jumlah bilangan ganjil) selalu sama dengan sisi kanan (𝑛^2). Seluruh hasil yang ditampilkan pada kolom Valid? bernilai True, artinya pernyataan tersebut selalu benar.

Dengan demikian, dapat disimpulkan bahwa jumlah 𝑛 bilangan ganjil pertama selalu menghasilkan kuadrat dari 𝑛. Pola ini berlaku umum, tidak hanya untuk
𝑛 sampai 15 saja, melainkan untuk semua bilangan bulat 𝑛≥0.

# Memeriksa soal persamaan (3):
1^2+2^2+3^2+⋯+n^2= $\frac{n(n+1)(2n+1)}{6}$


Berdasarkan tabel hasil perhitungan, terlihat bahwa untuk setiap nilai 𝑛 dari 1 sampai 15, jumlah kuadrat bilangan bulat pertama (𝐿𝐻𝑆) selalu sama dengan hasil rumus pada sisi kanan (𝑅𝐻𝑆). Semua hasil pengujian pada kolom Valid? menunjukkan nilai True.

Dengan demikian, dapat disimpulkan bahwa rumus jumlah kuadrat bilangan bulat positif dari 1 hingga 𝑛 selalu benar. Rumus ini tidak hanya berlaku untuk
𝑛 dari 1 sampai 15, tetapi juga untuk semua bilangan bulat 𝑛≥0.​
